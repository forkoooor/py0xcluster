```python
uni2_pools = Pools('univ2')
pools_summary, pools_df = uni2_pools.get_pools(
    min_daily_volume = 100000,
    min_daily_txns = 200,
    min_txns_per_token = 50000,
    start_date = (2022,10,10),
    end_date = (2022,11,10)
    )
pools_summary.head(50)
```

```python
pools_summary[(pools_summary['token0.txCount'] > 50000) & (pools_summary['token1.txCount'] > 50000)].head(50)
```

```python
pools_summary.dtypes
```

```python
pool_summary = pools_df.groupby('pairAddress').agg(
    {'dailyTxns': np.mean, 
    'dailyVolumeUSD':np.mean, 
    'token0.totalLiquidity': np.mean,
    'token1.totalLiquidity': np.mean,
    'token0.txCount' : np.mean,
    'token1.txCount' : np.mean
    })
```

```python
pool_id_cols = ['pairAddress','token0.symbol','token1.symbol','token0.id','token1.id']
pools_id_df = pools_df[pool_id_cols].drop_duplicates('pairAddress')
pool_summary.merge(pools_id_df, how='left', on='pairAddress').sort_values('token0.txCount', ascending=False).head(50)
```
