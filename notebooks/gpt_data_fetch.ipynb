```python
from gql import gql, Client
from gql.transport.requests import RequestsHTTPTransport

# Set up the client
transport = RequestsHTTPTransport(
    url='https://api.thegraph.com/subgraphs/name/messari/uniswap-v3-ethereum',
    use_json=True,
)
client = Client(transport=transport, fetch_schema_from_transport=True)

# Execute the query
value = 100000
query = gql('''
    query GetPoolsWithHighVolume($value: BigDecimal!) {
      liquidityPoolDailySnapshots(
        where: {
          dailyVolumeUSD_gte: $value
        }
      ) {
        id
        name
      }
    }
''')
variables = {'value': value}
result = client.execute(query, variable_values=variables)

# Print the IDs of the pools
pool_ids = [pool['id'] for pool in result['liquidityPoolDailySnapshots']]
pool_names = [pool['name'] for pool in result['liquidityPoolDailySnapshots']]
print(pool_ids)
```
